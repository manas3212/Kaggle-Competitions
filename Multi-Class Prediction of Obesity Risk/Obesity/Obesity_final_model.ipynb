```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_data = pd.read_csv("/content/drive/MyDrive/DL/Kaggle_comp/Obesity/train_processed.csv")
test_data = pd.read_csv("/content/drive/MyDrive/DL/Kaggle_comp/Obesity/test_processed.csv")
train_data
```

```python
test_data
```

```python
train_data=pd.get_dummies(data=train_data, columns=['NObeyesdad'])
```

```python
train_data[0:15]
```

```python
output=['NObeyesdad_0',	'NObeyesdad_1',	'NObeyesdad_2',	'NObeyesdad_3',	'NObeyesdad_4',	'NObeyesdad_5',	'NObeyesdad_6']
```

```python
x_train= train_data.drop(output,axis=1)
y_train=train_data[output]
x_test= test_data.drop(output,axis=1)
y_test=test_data[output]
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_test
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from keras import callbacks

# early_stopping = callbacks.EarlyStopping(
#     min_delta=0.001, # minimium amount of change to count as an improvement
#     patience=20, # how many epochs to wait before stopping
#     restore_best_weights=True,
# )

model = Sequential()
model.add(Dense(45, activation='relu', input_shape=(30,)))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(15, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(7, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(x_train, y_train, epochs=30,validation_split=0.2,batch_size = 256)
```

```python
model.evaluate(x_train,y_train,batch_size =128)
```

```python
yp=model.predict(x_test)
```

#creating output file in format thesubmission is required


```python
#ye got prob by softmak for each test example in yp, now with help of below func we get the max amoungt the probs which is our output class
out=[]
for n in range(len(x_test)):
  output = np.argmax(yp[n])
  out.append(output)
```

```python
# Define a dictionary mapping integer values to labels
labels = {0: "Insufficient_Weight", 1: "Normal_Weight", 2: "Obesity_Type_I", 3: "Obesity_Type_II", 4: "Obesity_Type_III", 5: "Overweight_Level_I", 6: "Overweight_Level_II"}

# Create an empty list to store the labels
labeled_array = []

# Iterate through the original array and look up the corresponding label in the dictionary
for value in out:
    labeled_array.append(labels.get(value))


```

```python
import pandas as pd


# Create a list of indices starting from 20759
indices = range(20758, 20758 + len(labeled_array))

# Create a DataFrame
df = pd.DataFrame({'data': labeled_array}, index=indices)

# Print the DataFrame
print(df),df.shape
```
